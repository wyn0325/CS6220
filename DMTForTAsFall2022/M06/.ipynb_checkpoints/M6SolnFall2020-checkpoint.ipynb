```python
import csv
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.datasets import load_boston

import matplotlib.pyplot as plt
```

### Load Dataset

```python
X, y = load_boston(return_X_y=True)
```

```python
features = ['CRIM per capita crime rate by town', 
'ZN proportion of residential land zoned for lots over 25,000 sq.ft.',
'INDUS proportion of non-retail business acres per town',
'CHAS Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)',
'NOX nitric oxides concentration (parts per 10 million)',
'RM average number of rooms per dwelling',
'AGE proportion of owner-occupied units built prior to 1940',
'DIS weighted distances to five Boston employment centres',
'RAD index of accessibility to radial highways',
'TAX full-value property-tax rate per $10,000',
'PTRATIO pupil-teacher ratio by town',
'B 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town',
'LSTAT % lower status of the population',
'Median value of owner-occupied homes in $1000’s']
```

### Split Dataset - Train 80%, Test 20%

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

### Linear Regression and Evaluation Metrics

```python
def apply_lr(X_train, X_test, y_train, y_test):
    lr = LinearRegression()
    lr.fit(X_train, y_train)
    y_pred = lr.predict(X_test)
    
    return lr.coef_, np.mean(((lr.predict(X_test) - y_test) ** 2)), lr.score(X_test, y_test), lr
```

```python
coef, mean_sq, var, _ = apply_lr(X_train, X_test, y_train, y_test)
```

```python
print('Coefficients: ', coef)
print('Mean Squared Error: ', mean_sq)
print('Variance Score: ', var)
```

### Linear Regression using Single Feature

```python
for i in range(X.shape[1]):
    print('--------------------------------------------------------------------')
    print('Selected features: ', features[i], '('+str(i+1)+')', '\nTarget: ', features[-1])

    coef, mean_sq, var, lr = apply_lr(X_train[:, [i]], X_test[:, [i]], y_train, y_test)
    
    x_min, x_max = X[:, [i]].min(), X[:, [i]].max()
    y_min = lr.predict([[x_min]])
    y_max = lr.predict([[x_max]])

    plt.plot([x_min, x_max], [y_min, y_max], '.b-')
    plt.scatter(X_train[:, [i]], y_train, c='cyan', alpha=0.5)
    plt.scatter(X_test[:, [i]], y_test, c='r', marker='^', alpha=0.8)
    
    plt.xlabel(features[i])
    plt.ylabel(features[-1])
    plt.show()
    
    print('Coefficients: ', coef)
    print('Mean squared error: ', mean_sq)
    print('Variance score: ', var)
    print('--------------------------------------------------------------------')
```

### Average over Evaluation Metrics

```python
scores = [0]*(X.shape[1]+1)
mean_sqs = [0]*(X.shape[1]+1)
vars = [0]*(X.shape[1]+1)

times = 10
for t in range(times):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
    score, mean_sq, var, _ = apply_lr(X_train, X_test, y_train, y_test)
    mean_sqs[0] += mean_sq
    vars[0] += var
    
    for i in range(X.shape[1]):
        score, mean_sq, var, _ = apply_lr(X_train[:, [i]], X_test[:, [i]], y_train, y_test)
        mean_sqs[i+1] += mean_sq
        vars[i+1] += var
```

```python
mean_sqs = np.array(mean_sqs)/times
vars = np.array(vars)/times
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,4))
ax[0].plot(mean_sqs, '*-')
ax[1].plot(vars, '*-')

ax[0].set(xlabel='features (0 is all features)', ylabel='Average Mean Sq Error')
ax[1].set(xlabel='features (0 is all features)', ylabel='Average Variance Score')

plt.show()
```

### Question Answers

    1. Based upon the linear models for the single features, feature-6 (RM average number of rooms per dwelling), and feature-13 (LSTAT % lower status of the population) appear to be most predictive for the target feature. Also this can be easily seen from the average evaluation metrics plots. 

    2. The best two features to predict the target feature are feature-6 and feature-13. So this two features are more likely to be selected, though it can not be said for sure that these two features will be best compared to any two features. But it is likely that at least feature-6 or feature-13 would be selected. 
    
    To see this experimentally, linear regression models are evaluated below for each pair of features. And we can see that, all the top models (with score > 0.6) have either feature-6 or feature-13. 

```python
print('Features: Evaluation Metrics')
for i in range(X.shape[1]):
    for j in range(i+1, X.shape[1]):
        score, mean_sq, var, _ = apply_lr(X_train[:, [i,j]], X_test[:, [i,j]], y_train, y_test)
        if score > 0.6:
            print(i+1, j+1, '\t: ', '%2f' %score, '%2f' %mean_sq, '%2f' %var)
```

    3. From the average evaluation metrics, we see that the model using all the features have maximum score, minimum mean squared error and maximum variance score. After that, feature-13 and feature-6 have better values compared to other single feature models.
